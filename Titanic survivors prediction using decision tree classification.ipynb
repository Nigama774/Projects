# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the training dataset

```python
dataset_train = pd.read_csv('train.csv')
X_train = dataset_train.iloc[: , 1:].values
Y_train = dataset_train.iloc[:, 0].values
print(X_train)
print(X_train[4])
```

# Importing the test dataset

```python
dataset_test = pd.read_csv('test.csv')
X_test = dataset_test.iloc[:, :].values
print(X_test)
```

# Encoding Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
X_train = np.array(ct.fit_transform(X_train))
ct_1 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
X_test = np.array(ct_1.fit_transform(X_test))
```

```python
print(X_train)
print(X_test)
```

# Encoding Categorical Data with Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train[:,4] = le.fit_transform(X_train[:,4])
```

```python
print(X_train[:,4])
```

```python
le_1 = LabelEncoder()
X_test[:,4] = le_1.fit_transform(X_test[:,4])
```

```python
print(X_test[:,4])
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Training the decision tree on training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, Y_train)
```

# Predicting the survivors

```python
y_pred = classifier.predict(X_test)
print(len(y_pred))
print(y_pred)
```
