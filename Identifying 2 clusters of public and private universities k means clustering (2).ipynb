# Importing the libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('College.csv')
X = dataset.iloc[:, 2:].values
y_real = dataset.iloc[:, [1]].values

```

# Encoding the y-real variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_real = le.fit_transform(y_real.ravel())
```

```python
print(y_real)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

# Training the k means model on the dataset

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, init = 'k-means++', n_init=50, max_iter=500, random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

## Aligning Cluster Labels with true labels

```python
from scipy.stats import mode
labels = np.zeros_like(y_kmeans)
for i in range(2):
  mask = (y_kmeans == i)
  labels[mask] = mode(y_real[mask])[0]
```

# Making confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_real, y_kmeans)
print(cm)
accuracy_score(y_real, y_kmeans)
```
