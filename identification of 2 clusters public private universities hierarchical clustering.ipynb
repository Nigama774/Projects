# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('College.csv')
X = dataset.iloc[:, 2:].values
y_real = dataset.iloc[:, [1]].values
```

# Encoding the y_real variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_real = le.fit_transform(y_real)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

# Training the hierarchical model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, linkage = 'ward')
y_hc = hc.fit_predict(X)
```

# Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_real, y_hc)
print(cm)
accuracy_score(y_real, y_hc)
```
